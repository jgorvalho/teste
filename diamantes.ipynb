**Diamantes: preços e características**

**Fonte: https://www.kaggle.com/shivam2503/diamonds**

```python
import pandas as pd
diamonds = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv')
diamonds
```

**para mostrar apenas o peso do quilate >= que 0,70**

```python
booleans = []
for w in diamonds.carat:
   if w >= .70:
       booleans.append(True)
   else:
       booleans.append(False)
#print(booleans)
diamonds.iloc[booleans]
```

**diamantes com comprimento>5 e largura>5 e profundidade>5**

```python
result = diamonds[(diamonds.x>5) & (diamonds.y>5) & (diamonds.z>5)]
result
```

**diamantes que são Premium ou Ideal, no campo cut**

```python
result = diamonds[(diamonds.cut == 'Premium') | (diamonds.cut == 'Ideal')]
result
```

**diamantes que estão com um Fair ou Good ou Premium, no campo cut**

```python
result = diamonds[diamonds.cut.isin(['Fair', 'Good', 'Premium'])]
result
```

```python
result=diamonds.mean()
result
```

```python
result=diamonds.mean(axis=1)
result.head()
```

```python
#result=diamonds.groupby('cut').price.mean()

resultCut=diamonds.groupby('cut')
result=resultCut.price.mean()
result
```

```python
result=diamonds.groupby('cut').price.count()
result
```

```python
result=diamonds.groupby('cut').price.min()
result
```

```python
result=diamonds.groupby('cut').price.max()
result
```

```python
result=diamonds.groupby('cut').price.agg(['count', 'min', 'max'])
result
```
