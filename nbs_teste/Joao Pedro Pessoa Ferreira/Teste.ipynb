### Q1 - Efectuar o parsing do campo measurement_time para datetime

```python
df_ms = pd.read_csv("measurements.csv", parse_dates=['measurement_time'])
df_ms.head(5)
```

```python
df_ms.dtypes
```

### Q2 - Retirar definitivamente (drop) os campos unnamed* do dataframe

```python
df_ms.drop(["Unnamed: 10", "Unnamed: 11"], axis=1, inplace=True)
df_ms.head(5)
```

### Q3 - Ordenar o dataframe de forma ascendente pelo campo measurement_time

```python
df_ms = df_ms.sort_values(by=['measurement_time'], ascending=[True])
df_ms.head(10)
```

### Q4-Escreva a fun√ß√£o para obter estas estat√≠sticas descritivas (novo dataframe)

```python
result=df_ms.describe()
result
```

### Q5 - Qual o valor do desvio padr√£o (std) do campo temperature ? (guarde-o numa variavel)

```python
temperature_std=result.loc["std","temperature"]
temperature_std
```

### Q6 - Qual o valor da m√©dia (mean) do campo temperature ? (guarde-o numa variavel)

```python
pressure_mean=result.loc["mean","pressure"]
pressure_mean
```
