Efectuar o parsing do campo measurement_time para datetime

```python
df_measurements = pd.read_csv("measurements.csv", parse_dates=['measurement_time'])
df_measurements.head(3)
```

Retirar definitivamente (drop) os campos unnamed* do dataframe

```python
df_measurements.drop(["Unnamed: 10", "Unnamed: 11"], axis=1, inplace=True)
df_measurements.head(3)
```

Ordenar o dataframe de forma ascendente pelo campo measurement_time

```python
df_measurements = df_measurements.sort_values(by=['measurement_time'], ascending=[True])
df_measurements.head(10)
```

Escreva a função para obter estas estatísticas descritivas

```python
result=df_measurements.describe()
result
```

Qual o valor do desvio padrão (std) do campo temperature ?

```python
temperature_std=result.loc["std","temperature"]
temperature_std
```

Qual o valor da média (mean) do campo pressão ?

```python
pressure_mean=result.loc["mean","pressure"]
pressure_mean
```

escreva o código para obtermos esses registos

```python
result1 = df_measurements[(df_measurements.vibration_y>=7) & (df_measurements.temperature>temperature_std) & (df_measurements.pressure>pressure_mean)]
result1.head(10)
```

```python
df_failures = pd.read_csv("failures.csv", parse_dates=['failure_time'])
df_failures.head(3)
```

```python
df_failures = df_failures.sort_values(by=['failure_time'], ascending=[True])
df_failures.head()
```

Efectue o agupamento (groupby) pelo campo gadget_id e calcule (cont) a quantidade de ocorrência de cada tipo de gadget

```python
result2=df_failures.groupby(["gadget_id"]).count()
result2
```

Crie uma gráfico com os dados de objecto (groupby): barras, sem legenda e um título

```python
result2.plot.bar(title="Nº de falhas/gadget", legend=False)
```
