Q1 - Efectuar o parsing do campo measurement_time para datetime

```python
# Ver os os data types
data_measurements.dtypes
```

```python
data_measurements = pd.read_csv("measurements.csv", parse_dates=['measurement_time'])

data_measurements.head()
```

```python
# Confirmar se mudou o data type
data_measurements.dtypes
```

Q2 - Retirar definitivamente (drop) os campos unnamed* do dataframe (Unnamed: 10 e Unnamed: 11)

```python
data_measurements.drop(['Unnamed: 10', 'Unnamed: 11'], inplace=True, axis=1)

data_measurements.head()
```

Q3 - Ordenar o dataframe de forma ascendente pelo campo measurement_time

```python
data_measurements = data_measurements.sort_values(by=['measurement_time'], ascending=[True])

data_measurements.head()
```

Q4 - Escreva a função para obter estas estatísticas descritivas (novo dataframe)

```python
result=data_measurements.describe()

result
```

Q5 - Qual o valor do desvio padrão (std) do campo temperature ? (guarde-o numa variavel)

```python
desvio_temperature=result.loc['std','temperature']

desvio_temperature
```

Q6 - Qual o valor da média (mean) do campo pressure ? (guarde-o numa variavel)

```python
media_pressure=result.loc["mean","pressure"]

media_pressure
```

Q7 - Falha é um estado quando se verificam as condições (&)

vibration_y >= 7
temperature > que o valor std do campo temperature
pressure > que o valor mean do campo pressure


```python
result = data_measurements[(data_measurements.vibration_y>=7) & (data_measurements.temperature>desvio_temperature) & (data_measurements.pressure>media_pressure)]

result.head()
```

Entrada de outra tabela

```python
data_failures = pd.read_csv("failures.csv", parse_dates=['failure_time'])

data_failures.head()
```

```python
data_failures = data_failures.sort_values(by=['failure_time'], ascending=[True])

data_failures.head()
```

Q8 - Efectue o agupamento (groupby) pelo campo gadget_id e calcule (count) a quantidade de ocorrência de cada tipo de gadget

```python
result=data_failures.groupby(["gadget_id"]).count()

result
```

Q9 - Crie uma gráfico com os dados de objecto (groupby): barras, sem legenda e um título

```python
result.plot.bar(title="Nº de falhas/gadget", legend=False)

```
